## WRITE  A  PROGRAM  TO  SPLIT A  WORD  INTO  PAIRâ€™S  AT  ALL  POSSIBLE 
## POSITIONS. FOR EXAMPLE, CARRIED WILL BE SPLIT INTO {C, ARRIED, CA 
## ,RRIED, CAR, RIED, CARR, IED, CARRI, ED, CARRI, D}.

```python
# Cell 1: Split word into all possible pairs

def split_word_into_pairs(word: str):
    pairs = []
    for i in range(1, len(word)):
        pairs.append((word[:i], word[i:]))
    return pairs

# Example usage
word = "CARRIED"
pairs = split_word_into_pairs(word)

print("Original Word:", word)
print("All Possible Pairs:")
for p in pairs:
    print(p)
```

ðŸ“˜ Cell 2: Generate Prefixes and Suffixes

```python
# Cell 2: Generate all prefixes and suffixes

def generate_prefixes_suffixes(word: str):
    prefixes = [word[:i] for i in range(1, len(word) + 1)]
    suffixes = [word[i:] for i in range(len(word))]
    return prefixes, suffixes

# Example usage
word = "CARRIED"
prefixes, suffixes = generate_prefixes_suffixes(word)

print("Original Word:", word)
print("\nPrefixes:")
for p in prefixes:
    print(p)

print("\nSuffixes:")
for s in suffixes:
    print(s)
```

Randomized Split

# Cell 3: Randomized splits of the word



```python
import random

def random_splits(word: str):
    positions = list(range(1, len(word)))
    random.shuffle(positions)
    return [(word[:i], word[i:]) for i in positions]

# Example usage
word = "CARRIED"
splits = random_splits(word)

print("Original Word:", word)
print("Random Splits:")
for s in splits:
    print(s)
```
